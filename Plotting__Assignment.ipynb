```python
# YOUR CODE HERE

plt.figure(figsize=(10,6))
sns.histplot(df['co2'], bins=20, kde=True, color='orange')
plt.title('Distribution of CO2 Emissions')
plt.xlabel('CO2 Emissions (g/mi)')
plt.ylabel('Number of Vehicles')
plt.grid(True)
plt.savefig('histogram_co2_emissions.png')
plt.show()
```

```python
# YOUR CODE HERE

plt.figure(figsize=(10,6))
sns.histplot(df['comb'], bins=25, kde=True, color='green')
plt.title('Distribution of Combined Fuel Efficiency')
plt.xlabel('Combined Fuel Efficiency (MPG)')
plt.ylabel('Number of Vehicles')
plt.grid(True)
plt.savefig('histogram_combined_fuel_efficiency.png')
plt.show()
```

```python
# YOUR CODE HERE

plt.figure(figsize=(10,6))
sns.histplot(df['ghgScore'], bins=10, kde=False, color='blue')
plt.title('Distribution of Greenhouse Gas Score')
plt.xlabel('GHG Score')
plt.ylabel('Number of Vehicles')
plt.grid(True)
plt.savefig('histogram_ghg_score.png')
plt.show()

```

```python
# Select only numerical columns
numeric_df = df.select_dtypes(include=['float64', 'int64'])

# Compute correlation matrix
corr_matrix = numeric_df.corr()

# Plot heatmap
plt.figure(figsize=(12, 8))
sns.heatmap(corr_matrix, annot=True, cmap='RdBu_r', center=0)
plt.title('Correlation Heatmap of Fuel Efficiency Data')
plt.xticks(rotation=45)
plt.tight_layout()
plt.savefig('figures/heatmap.png')
plt.show()


```
